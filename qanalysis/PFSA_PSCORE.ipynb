```python
import pandas as pd
import pylab as plt
import seaborn as sns
sns.set()
PATH='/home/ishanu/comparae_05.csv'
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True,figsize=[5,6])

ax=axes[0]
df=pd.read_csv(PATH,header=None,names=['fpr','typ','cohort','label','out','num']).drop('fpr',axis=1)
dfa=df[df.cohort=='ALL'].drop('cohort',axis=1)
D1=dfa[dfa.typ=='raw'].drop('typ',axis=1).set_index(['label','out'])
D2=dfa[dfa.typ=='perturbed'].drop('typ',axis=1).set_index(['label','out'])

TYP='FN'
D1_=D1.reset_index()#.drop('label',axis=1)
D1_=D1_[D1_.out==TYP].drop('out',axis=1).set_index('label')
D2_=D2.reset_index()#.drop('label',axis=1)
D2_=D2_[D2_.out==TYP].drop('out',axis=1).set_index('label')
D1_.columns=[TYP+' raw']
D2_.columns=[TYP+' perturbed']

D=D1_.join(D2_)
a=D.min().min()*.8
D.plot(ax=ax,kind='bar',alpha=.8,ylim=[a,None])
#p1=ax.legend()
ax.legend(bbox_to_anchor=(1.5, 1.05))

ax=axes[1]
TYP='FP'
D1_=D1.reset_index()#.drop('label',axis=1)
D1_=D1_[D1_.out==TYP].drop('out',axis=1).set_index('label')
D2_=D2.reset_index()#.drop('label',axis=1)
D2_=D2_[D2_.out==TYP].drop('out',axis=1).set_index('label')
D1_.columns=[TYP+' raw']
D2_.columns=[TYP+' perturbed']

D=D1_.join(D2_)
a=D.min().min()*.5
D.plot(ax=ax,kind='bar',alpha=.8,ylim=[a,None])
ax.legend(bbox_to_anchor=(1.1, 1.05));
plt.savefig('full_cohort.png',dpi=300,bbox_inches='tight',transparent=True)
```
