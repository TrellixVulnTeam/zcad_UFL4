```python
import pandas as pd
import seaborn as sns
import pylab as plt
plt.style.use('fivethirtyeight')
```

```python
df0=pd.read_csv('nitems_SvS.csv')
df1=pd.read_csv('nitems_SvO.csv')
df2=pd.read_csv('nitems_SSvO.csv')
```

```python
sns.distplot(df0.numt)
sns.distplot(df1.numt)
ax=sns.distplot(df2.numt)
ax.set_xlabel('No. of items used')
ax.legend(['schizophrenia vs schizoaffective',
           'schizophrenia vs others',
           'schizophrenia + schizoaffective  vs others'],
          bbox_to_anchor=(0.1, 1.3), loc=2, borderaxespad=0.)
plt.savefig('items.pdf',dpi=300,bbox_inches='tight',transparent=True)
```
